# Reasoning dataset from raiting and text generation

### dataset geneartion


```python
#use this file to generate a training dataset from .txt files of ssts uc and output tables in markdown format

import os
import json
import re
import csv
import io
DATA_DIR="./data"#folder with ssts uc and table folders
OUT_NAME="./DatasetRTTReformed.jsonl"
roles={"sys":"system","usr":"user","mdl":"assistant"}
def sampl_to_jsonl_line(sample)->str: #add tokens and shield special symbold
    sampleText="<|begin_of_text|>"
    for msg in sample:
        for bad in ["begin_of_text","start_header_id","end_header_id","eot_id"]:
            msg["text"]=msg["text"].replace("<|"+bad+"|>",bad)#remove unwanted tokens
        msg["text"]=msg["text"].replace("\n","\\n") #remove "\n"
        sampleText+=f"<|start_header_id|>{roles[msg['role']]}<|end_header_id|>{msg['text']}<|eot_id|>"
    sample={"text":sampleText}
    sample=json.dumps(sample)
    return sample

def is_markdown_table_valid(markdown_table: str) -> bool:  #gptCode
    # Split the input table by lines
    lines = markdown_table.strip().split('\n')

    # Check if there are at least two lines: header and separator
    if len(lines) < 2:
        return False

    # Check if the second line is a valid separator
    header_line = lines[0].strip()
    separator_line = lines[1].strip()

    # Separator should contain only '-' and optional ':' for alignment
    if not re.match(r'^\|?(\s*-+\s*\|?)+$', separator_line):
        return False

    # The number of columns in the header line must match the number of separators
    num_columns = header_line.count('|') + 1
    num_separators = separator_line.count('|') + 1
    if num_columns != num_separators:
        return False

    # Now check the data rows
    for row in lines[2:]:
        row = row.strip()
        row_columns = row.count('|') + 1
        if row_columns != num_columns:
            return False

    return True

def parse_markdown_table(markdown_table: str): #gptCode
    # Split the input table by lines
    lines = markdown_table.strip().split('\n')

    # First line is header row
    header_line = lines[0].strip()
    headers = [col.strip() for col in header_line.split('|')[1:-1]]  # Remove leading and trailing pipes

    # The second line is the separator
    separator_line = lines[1].strip()

    # Ensure the separator is valid (contains - and optional alignment ':')
    if not re.match(r'^\|?(\s*-+\s*\|?)+$', separator_line):
        raise ValueError("Invalid Markdown table separator line.")

    # Now process data rows
    data_rows = []
    for row in lines[2:]:
        row = row.strip()
        columns = [col.strip() for col in row.split('|')[1:-1]]  # Remove leading and trailing pipes
        data_rows.append(columns)

    return (headers, data_rows)

def list_to_csv_line(data): #gptCode
    # Create an in-memory string buffer
    output = io.StringIO()

    # Create a CSV writer that writes to the buffer
    writer = csv.writer(output, quotechar='"', quoting=csv.QUOTE_MINIMAL)

    # Write the list as a single row in the CSV format
    writer.writerow(data)

    # Get the CSV string from the buffer and remove any trailing newline
    csv_line = output.getvalue().strip()

    # Close the buffer
    output.close()

    return csv_line

# load prompts
with open("./prompts/RTT/userPrompt.txt",encoding="utf-8") as F:
    userPrompt=F.read()
with open("./prompts/RTT/sysPrompt.txt",encoding="utf-8") as F:
    sysPrompt=F.read()
with open("./prompts/RTT/respPrompt.txt",encoding="utf-8") as F:
    respPrompt=F.read()

with open(DATA_DIR+"/complienceLevels.txt",encoding="utf-8") as F:
    complienceLevels=list(map(lambda x:x[2:-1] ,F.readlines()))#remove numbers in the begiging of the line
outF= open(OUT_NAME,"w",encoding="utf-8")

for fileName in os.listdir(DATA_DIR+"/uc/"):
    toReplace={"clNum":0,"clText":""}
    toReplace["clNum"]=int(fileName.split(".")[-2].split("_")[-1])
    toReplace["clText"]=complienceLevels[toReplace["clNum"]-1]
    buf={"uc":"","ssts":"", "table":""}

    for name in buf.keys():
        with open(DATA_DIR+f"/{name}/"+fileName,encoding="utf-8",errors="ignore") as F:
            for line in F.readlines():
                if len(line)>1:
                    buf[name]+=line
    if not is_markdown_table_valid(buf["table"]):
        print(f"Bad markdown table at {fileName}")
        exit()

    # from here assuming that data is clean

    #need to reformat mrkdown table to the csv string
    table=buf["table"]
    table=parse_markdown_table(table)[1]#extract only data
    difference=[table[i][0] for i in range(len(table))]
    quotation=[table[i][1] for i in range(len(table))]
    difference=" ".join(list([str(i+1)+".  "+difference[i] for i in range(len(difference))]))
    quotation=" ".join(list([str(i+1)+".  "+quotation[i] for i in range(len(quotation))]))
    table=[difference,quotation]
    csvString=list_to_csv_line(table)
    toReplace["markdownTable"]=buf["table"]
    toReplace["csvString"]=csvString

    userBuf=userPrompt
    sysBuf=sysPrompt
    respBuf=respPrompt

    for toRep,rep in (list(buf.items())+list(toReplace.items())):#inserting data in placeholders place
        userBuf=userBuf.replace("{"+toRep+"}",str(rep))
        sysBuf=sysPrompt.replace("{"+toRep+"}",str(rep))
        respBuf=respBuf.replace("{"+toRep+"}",str(rep))

    sample=[
        {"role":"sys","text":sysBuf},
        {"role":"usr","text":userBuf},
        {"role":"mdl","text":respBuf}
    ]

    text=sampl_to_jsonl_line(sample)
    outF.write(text+"\n")

outF.close()
```

### Reasoning from texts and raiting model exaple run


```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import re
import time
model_path = "Okun-Elya/Mistral-Nemo-Instruct-2407-ComplienceCheckerV4"
DATA_DIR="./data"#folder with ssts uc and table folders

def is_markdown_table_valid(markdown_table: str) -> bool:  #gptCode
    # Split the input table by lines
    lines = markdown_table.strip().split('\n')

    # Check if there are at least two lines: header and separator
    if len(lines) < 2:
        return False

    # Check if the second line is a valid separator
    header_line = lines[0].strip()
    separator_line = lines[1].strip()

    # Separator should contain only '-' and optional ':' for alignment
    if not re.match(r'^\|?(\s*-+\s*\|?)+$', separator_line):
        return False

    # The number of columns in the header line must match the number of separators
    num_columns = header_line.count('|') + 1
    num_separators = separator_line.count('|') + 1
    if num_columns != num_separators:
        return False

    # Now check the data rows
    for row in lines[2:]:
        row = row.strip()
        row_columns = row.count('|') + 1
        if row_columns != num_columns:
            return False

    return True

def parse_markdown_table(markdown_table: str): #gptCode
    # Split the input table by lines
    lines = markdown_table.strip().split('\n')

    # First line is header row
    header_line = lines[0].strip()
    headers = [col.strip() for col in header_line.split('|')[1:-1]]  # Remove leading and trailing pipes

    # The second line is the separator
    separator_line = lines[1].strip()

    # Ensure the separator is valid (contains - and optional alignment ':')
    if not re.match(r'^\|?(\s*-+\s*\|?)+$', separator_line):
        raise ValueError("Invalid Markdown table separator line.")

    # Now process data rows
    data_rows = []
    for row in lines[2:]:
        row = row.strip()
        columns = [col.strip() for col in row.split('|')[1:-1]]  # Remove leading and trailing pipes
        data_rows.append(columns)

    return (headers, data_rows)
PROMPTS_DIR="./prompts"

# load prompts (RTT)
with open(PROMPTS_DIR+"/RTT/sysPrompt.txt",encoding="utf-8") as F:
    RTTsysPrompt=F.read()
with open(PROMPTS_DIR+"/RTT/userPrompt.txt",encoding="utf-8") as F:
    RTTuserPrompt=F.read()

with open(DATA_DIR+"/complienceLevels.txt",encoding="utf-8") as F:
    complienceLevels=list(map(lambda x:x[2:-1] ,F.readlines()))#remove numbers in the begiging of the line

#load model
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForCausalLM.from_pretrained(
    model_path,
    device_map="auto",
    torch_dtype='auto',
).eval()


def getReasoningTable(uc:str,ssts:str,complienceRating:int):
    ''' Generates reasoning and quotes differences in texts using  ComplienceRaiting that variates from 1 to 4'''
    global RTTuserPrompt, RTTsysPrompt,tokenizer,model
    sysPromptBuf=RTTsysPrompt
    userPromptBuf=RTTuserPrompt

    toReplace={"clNum":0,"clText":"","uc":uc,"ssts":ssts}
    toReplace["clNum"]=complienceRating
    toReplace["clText"]=complienceLevels[toReplace["clNum"]]

    for toRep,rep in list(toReplace.items()):#inserting data in placeholders place
        sysPromptBuf=sysPromptBuf.replace("{"+toRep+"}",str(rep))
        userPromptBuf=userPromptBuf.replace("{"+toRep+"}",str(rep))

    messages = [
        {"role": "system", "content": sysPromptBuf},
        {"role": "user", "content": userPromptBuf}
    ]

    difference=""
    quotation=""
    tryCounter=0
    while tryCounter<2:
        input_ids = tokenizer.apply_chat_template(conversation=messages, tokenize=True, add_generation_prompt=True, return_tensors='pt')
        output_ids = model.generate(input_ids.to('cuda'),max_new_tokens=10000)
        response = tokenizer.decode(output_ids[0][input_ids.shape[1]:], skip_special_tokens=True)
        try:
            table=parse_markdown_table(response)[1]
            difference=[table[i][0] for i in range(len(table))]
            quotation=[table[i][1] for i in range(len(table))]
            difference=" ".join(list([str(i+1)+".  "+difference[i] for i in range(len(difference))]))
            quotation=" ".join(list([str(i+1)+".  "+quotation[i] for i in range(len(quotation))]))
            break
        except Exception as E:
            print(E.with_traceback())
            # most probable is that model was unable to generate an adequate table

            messages.append({"role": "system", "content": "Respond ONLY in markdown table as was provided in the example "})
        tryCounter+=1
    return (difference,quotation)

```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import re
import time
model_path = "Okun-Elya/Mistral-Nemo-Instruct-2407-CCFCV1"
DATA_DIR="./data"#folder with ssts uc and table folders

def is_markdown_table_valid(markdown_table: str) -> bool:  #gptCode
    # Split the input table by lines
    lines = markdown_table.strip().split('\n')

    # Check if there are at least two lines: header and separator
    if len(lines) < 2:
        return False

    # Check if the second line is a valid separator
    header_line = lines[0].strip()
    separator_line = lines[1].strip()

    # Separator should contain only '-' and optional ':' for alignment
    if not re.match(r'^\|?(\s*-+\s*\|?)+$', separator_line):
        return False

    # The number of columns in the header line must match the number of separators
    num_columns = header_line.count('|') + 1
    num_separators = separator_line.count('|') + 1
    if num_columns != num_separators:
        return False

    # Now check the data rows
    for row in lines[2:]:
        row = row.strip()
        row_columns = row.count('|') + 1
        if row_columns != num_columns:
            return False

    return True

def parse_markdown_table(markdown_table: str): #gptCode
    # Split the input table by lines
    lines = markdown_table.strip().split('\n')

    # First line is header row
    header_line = lines[0].strip()
    headers = [col.strip() for col in header_line.split('|')[1:-1]]  # Remove leading and trailing pipes

    # The second line is the separator
    separator_line = lines[1].strip()

    # Ensure the separator is valid (contains - and optional alignment ':')
    if not re.match(r'^\|?(\s*-+\s*\|?)+$', separator_line):
        raise ValueError("Invalid Markdown table separator line.")

    # Now process data rows
    data_rows = []
    for row in lines[2:]:
        row = row.strip()
        columns = [col.strip() for col in row.split('|')[1:-1]]  # Remove leading and trailing pipes
        data_rows.append(columns)

    return (headers, data_rows)
PROMPTS_DIR="./prompts"

# load prompts (ALL)
with open(PROMPTS_DIR+"/ALL/sysPrompt.txt",encoding="utf-8") as F:
    ALLsysPrompt=F.read()
with open(PROMPTS_DIR+"/ALL/userPrompt1.txt",encoding="utf-8") as F:
    ALLuserPrompt1=F.read()
with open(PROMPTS_DIR+"/ALL/userPrompt2.txt",encoding="utf-8") as F:
    ALLuserPrompt2=F.read()


with open(DATA_DIR+"/complienceLevels.txt",encoding="utf-8") as F:
    complienceLevels=list(map(lambda x:x[2:-1] ,F.readlines()))#remove numbers in the begiging of the line

#load model
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForCausalLM.from_pretrained(
    model_path,
    device_map="auto",
    torch_dtype='auto',
).eval()

def getRaitingAndTable(uc:str,ssts:str):
    ''' Generates raiting,reasoning and quotes differences in texts  from 2 texts '''
    global ALLsysPrompt, ALLuserPrompt1,ALLuserPrompt2,tokenizer,model
    sysPromptBuf=ALLsysPrompt
    userPrompt1Buf=ALLuserPrompt1
    userPrompt2Buf=ALLuserPrompt2

    toReplace={"uc":uc,"ssts":ssts}

    for toRep,rep in list(toReplace.items()):#inserting data in placeholders place
        sysPromptBuf=sysPromptBuf.replace("{"+toRep+"}",str(rep))
        userPrompt1Buf=userPrompt1Buf.replace("{"+toRep+"}",str(rep))
        userPrompt2Buf=userPrompt2Buf.replace("{"+toRep+"}",str(rep))

    messages = [
        {"role": "system", "content": sysPromptBuf},
        {"role": "user", "content": userPrompt1Buf}
    ]# this is request for complaince raiting number


    tryCounter=0
    raiting =3# right in the middle, in case of the failure good for mse
    #extract raiting
    while tryCounter<2:
        input_ids = tokenizer.apply_chat_template(conversation=messages, tokenize=True, add_generation_prompt=True, return_tensors='pt')
        output_ids = model.generate(input_ids.to('cuda'),max_new_tokens=10000)
        response = tokenizer.decode(output_ids[0][input_ids.shape[1]:], skip_special_tokens=True)
        try:
            response=response.strip()
            response=int(response)
            raiting=response
            break
        except Exception as E:
            print(E.with_traceback())
            # most probable is that model was unable to generate an adequate table
            messages.append({"role": "system", "content": "Respond ONLY with one number"})
        tryCounter+=1
    messages.append({"role": "assistant", "content": str(raiting)})
    messages.append({"role": "user", "content": userPrompt2Buf})

    #extract markdown table
    difference=""
    quotation=""
    tryCounter=0
    while tryCounter<2:
        input_ids = tokenizer.apply_chat_template(conversation=messages, tokenize=True, add_generation_prompt=True, return_tensors='pt')
        output_ids = model.generate(input_ids.to('cuda'),max_new_tokens=10000)
        response = tokenizer.decode(output_ids[0][input_ids.shape[1]:], skip_special_tokens=True)
        try:
            table=parse_markdown_table(response)[1]
            difference=[table[i][0] for i in range(len(table))]
            quotation=[table[i][1] for i in range(len(table))]
            difference=" ".join(list([str(i+1)+".  "+difference[i] for i in range(len(difference))]))
            quotation=" ".join(list([str(i+1)+".  "+quotation[i] for i in range(len(quotation))]))
            break
        except Exception as E:
            print(E.with_traceback())
            # most probable is that model was unable to generate an adequate table

            messages.append({"role": "system", "content": "Respond ONLY in markdown table as was provided in the example "})
        tryCounter+=1
    return (raiting,difference,quotation)

if __name__ == "__main__":
    print("Started!")

    with open(DATA_DIR+"/ssts/1_2.txt",encoding="utf-8",errors="ignore") as F:
        ssts=F.read()
    with open(DATA_DIR+"/uc/1_2.txt",encoding="utf-8",errors="ignore") as F:
        uc=F.read()
    genTime=time.time()

    resp=getRaitingAndTable(uc,ssts)
    genTime=time.time()-genTime
    print(f"Generation took {int(genTime*1000)}ms")
    (raiting,difference,quotation)=resp
    print(f"Raiting:{raiting}({complienceLevels[max(min(0,raiting-1),len(complienceLevels)-1)]})")
    print(f"Difference:{difference}")
    print(f"Quotation:{quotation}")

    print(resp)
```
